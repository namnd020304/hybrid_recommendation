```python
import pandas as pd
import json
import re
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("E:\\Project2\\code\\data1Gb\\movies.csv", sep = ',')
```

```python
movies
```

```python
movies['genres'] = movies.genres.str.split('|')
```

```python
movies
```

```python
movies["title"]=movies.title.str.replace('\(\d\d\d\d\)', '', regex=True)
```

```python
movies
```

```python
movies["genres"] = movies["genres"].apply(
    lambda x: ["Drama"] if x == ["(no genres listed)"] else x
)
```

```python
movies
```

```python
for index, row in movies.iterrows():
    for genre in row.genres:
        movies.at[index, genre] = 1
movies
```

```python
movies = movies.fillna(0)
movies
```

```python
movies = movies.drop(['title','genres'], axis=1)
movies
```

```python
rating = pd.read_csv("E:\Project2\code\data1Gb\\ratings.csv")
rating
```

```python
vy1 = rating[rating["userId"] == 1]
vy1 =vy1.drop("timestamp", axis=1)
vy1
```

```python
user_profile = pd.merge(vy1, movies, how='left', on=['movieId', 'movieId'])
user_profile
```

```python
genre_col = user_profile.columns[3:]
list(genre_col)
```

```python
# Cách 1: Nhân trực tiếp các cột
for genre in genre_col:
    user_profile[genre] = user_profile[genre] * user_profile['rating']

user_profile
```

```python
user_list = list(user_profile.sum()[3:])
sum_user_list = sum(user_list)
cal_user = pd.Series([genre/sum_user_list for genre in user_list], index=genre_col)
```

```python
cal_user
```

```python
movies_info = pd.read_csv("E:\\Project2\\code\\data1Gb\\movies.csv", sep=',')

# Tính similarity score (chỉ dùng các cột genre để tính, không dùng movieId)
genre_columns = movies.columns[1:]  # Lấy tất cả cột trừ movieId
movies['similarity_score'] = movies[genre_columns].dot(cal_user)

# Sort theo similarity score giảm dần
movies_sorted = movies.sort_values('similarity_score', ascending=False)

# Merge với movies_info để lấy tên phim
top_10 = movies_sorted[['movieId', 'similarity_score']].head(10)
top_10_with_title = pd.merge(top_10, movies_info[['movieId', 'title']], on='movieId', how='left')

# Hiển thị kết quả
print("Top 10 phim được recommend cho user 1:")
print(top_10_with_title[['movieId', 'title', 'similarity_score']])
```
